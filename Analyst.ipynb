```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('bronze.csv')
```

```python
df['date_heure'] = pd.to_datetime(df['date_heure'])

df.info()
```

```python
df.set_index('date_heure', inplace=True) # Définition de 'date_heure' comme index
```

## 2. Affichez cette même série temporelle en rassemblant:
- Par année
- Par trimestre
- Par mois
- Par semaine
- Par jour


```python
def resample_and_plot(df, frequencies, column_name):
    for freq, label in frequencies:
        # Rééchantillonnage des données selon la fréquence spécifiée
        df_resampled = df.resample(freq).mean()
        print(df_resampled)
        # Création d'un graphique pour afficher la série temporelle rééchantillonnée
        plt.figure(figsize=(10, 6))
        # met les markor en rouge et la ligne en bleu
        plt.plot(df_resampled.index, df_resampled[column_name], 'ro', markersize=3, color='red')
        plt.plot(df_resampled.index, df_resampled[column_name], linewidth=1, color='blue')
        plt.title(f'Consommation Brute d\'Électricité par {label} Hauts-de-France')
        plt.xlabel('Date')
        plt.ylabel('Consommation Brute d\'Électricité (RTE)')
        plt.grid(True)
        plt.tight_layout()
        plt.show()

# Utilisation de la fonction avec une liste de tuples contenant des fréquences et des étiquettes
frequencies = [('Y', 'Année'), ('Q', 'Trimestre'), ('M', 'Mois'), ('W', 'Semaine'), ('D', 'Jour')]
resample_and_plot(df, frequencies, 'consommation_brute_electricite_rte')
```

## 3. A partir de la time série agrégée à la journée, créez une moyenne mobile sur une semaine, centrée et symétrique.


```python
df_daily = df.resample('D').mean()
df_daily['moyenne_mobile'] = df_daily['consommation_brute_electricite_rte'].rolling(window=7, center=True).mean()
plt.figure(figsize=(10, 6))
plt.plot(df_daily.index, df_daily['consommation_brute_electricite_rte'], label='Consommation Journalière')
plt.plot(df_daily.index, df_daily['moyenne_mobile'], label='Moyenne Mobile sur une Semaine', color='red')
plt.title('Consommation Brute d\'Électricité avec Moyenne Mobile Hauts-de-France')
plt.xlabel('Date')
plt.ylabel('Consommation Brute d\'Électricité (RTE)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

df_daily
```

## 4.Quelle est la différence entre la moyenne mobile et les agrégations précédentes?


Dans notre analyse, nous avons d'abord utilisé différentes méthodes d'agrégation, comme l'agrégation annuelle, trimestrielle, mensuelle, hebdomadaire, et quotidienne, pour rassembler les données dans des blocs de temps distincts et calculer une statistique (comme la moyenne) pour chaque bloc. Cela nous donne une vision claire des tendances à plus long terme, mais nous pourrions perdre des détails nuancés qui sont perceptibles dans des intervalles de temps plus courts.

En revanche, la moyenne mobile, en particulier une centrée et symétrique, ne segmente pas les données en blocs séparés, mais utilise plutôt une fenêtre glissante pour calculer une moyenne en continu. Dans notre cas, chaque point dans la moyenne mobile est calculé à partir de la moyenne des 3 jours précédents, du jour même, et des 3 jours suivants, nous permettant ainsi de voir des fluctuations plus subtiles. C'est une technique précieuse pour observer les tendances et les modèles à court terme, tout en lissant le bruit quotidien, facilitant ainsi l'identification de tendances sous-jacentes plus significatives dans notre série temporelle.

## 5. A l’aide de box-plots, observez la saisonnalité:
- trimestrielle (chaque trimestre)
- mensuelle (chaque mois)
- hebdomadaire (chaque jour)
- journalière (chaque heure)


```python
df_boxplot = df.reset_index()
df_boxplot['trimestre'] = df_boxplot['date_heure'].dt.to_period('Q')

# Tri des valeurs par trimestre
df_boxplot = df_boxplot.sort_values('trimestre')
# Création des boxplots
plt.figure(figsize=(12,6))
sns.boxplot(x='trimestre', y='consommation_brute_electricite_rte', data=df_boxplot)
plt.xticks(rotation=90)
plt.title('Saisonnalité Trimestrielle')
plt.ylabel('Consommation Brute d\'Électricité (RTE)')
plt.xlabel('Trimestre')
plt.tight_layout()
plt.show()
```
